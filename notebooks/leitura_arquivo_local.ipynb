# Análise dos dados Utilizando Arquivo de forma Local

A premissa deste notebook é realizar a análise dos dados com a base armazenada localmente garantindo segurança de manuseio e redução de custos de acesso ao bucket S3.

### Importação de Bibliotecas

```python
import duckdb
import pandas as pd
from dotenv import load_dotenv
import os
```

### Carregando Variáveis de Ambiente

Para garantir a privacidade e segurança do ambiente de trabalho local, o caminho de acesso ao arquivo foi definido no .env, evitando possível vazamento de informações indesejadas no repositório.

- Caso você queira testar adicione essa linha no seu arquivo.env 
~~~
ARQUIVO_LOCAL = "C:\Pasta\Do\Seu\Arquivo\"
~~~
- Baixe o arquivo através desse link: [Portal de Dados TSE - Comparecimento 2022]( https://cdn.tse.jus.br/estatistica/sead/odsele/perfil_comparecimento_abstencao/perfil_comparecimento_abstencao_2022.zip);
- Extraia o CSV de Interesse.


```python
load_dotenv()
caminho_arquivo = os.getenv("ARQUIVO_LOCAL")
```

Definição das variáveis de acesso ao arquivos e banco de dados.

```python
file = f"{caminho_arquivo}perfil_comparecimento_abstencao_2022_BRASIL.csv" # Altere o nome do arquivo caso necessário.
db_file = 'eleicoes.db' # Nome do Banco de Dados Local para armazenamento - DuckDB
```

```python
comp = f"read_csv('{file}', sep=';',encoding='latin-1',all_varchar=True)"
```

### Definindo Conexão e Criando Banco Local 

```python
con = duckdb.connect(db_file)
```

### Criando tabela da base de comparecimento para o Banco de Dados.

```python
con.execute(f"""
    CREATE TABLE IF NOT EXISTS
    comparecimento_eleicoes AS 
    SELECT * FROM {comp}
""")
```

### Executando Consultas na Tabela

```python
con.sql("SELECT * FROM comparecimento_eleicoes LIMIT 10").df()
```

### Fechando Conexão com o Banco de Dados

```python
con.close()
```
